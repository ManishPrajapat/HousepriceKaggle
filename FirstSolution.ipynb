```python
#Dropping price from train
train_data.drop('SalePrice',axis=1,inplace=True)
```

```python
#Getting combined dataset
def get_combined():
    global train_data,test_data
    combined = train_data.append(test_data)
    combined.drop('Id',axis=1,inplace=True)
    combined.reset_index()
    return combined
combined = get_combined()
```

```python
#Analysing combined
combined.count()
```

```python
#Processing MSSubClass
def process_mssubclass():
    global combined
    combined.MSSubClass = (combined.MSSubClass - combined.MSSubClass.min())/(combined.MSSubClass.max()-combined.MSSubClass.min())
process_mssubclass()
```

```python
combined.MSZoning.fillna(combined.MSZoning.mode(),inplace=True)
```

```python
#Processing MSZoning
def process_MSZoning():
    global combined
    dummydata = pd.get_dummies(combined.MSZoning,prefix='MSZoning')
    combined = pd.concat([combined,dummydata],axis=1)
    combined.drop('MSZoning',axis=1,inplace=True)
process_MSZoning()
```

```python
combined.head()
```

```python
combined.LotFrontage.fillna(combined.LotFrontage.mean(),inplace=True)
#Processing LotFrontage
def process_LotFrontage():
    global combined
    combined.LotFrontage = (combined.LotFrontage - combined.LotFrontage.min())/(combined.LotFrontage.max()-combined.LotFrontage.min())
process_LotFrontage()
```

```python
#Processing LotArea
def process_LotArea():
    global combined
    combined.LotArea = (combined.LotArea - combined.LotArea.min())/(combined.LotArea.max()-combined.LotArea.min())
process_LotArea()
```

```python
#Processing Street
def process_Street():
    global combined
    dummydata = pd.get_dummies(combined.Street,prefix='Street')
    combined = pd.concat([combined,dummydata],axis=1)
    combined.drop('Street',axis=1,inplace=True)
process_Street()
```

```python
combined.Alley.fillna('UNKNOWN',inplace=True)
#Processing Alley
def process_Alley():
    global combined
    dummydata = pd.get_dummies(combined.Alley,prefix='Alley')
    combined = pd.concat([combined,dummydata],axis=1)
    combined.drop('Alley',axis=1,inplace=True)
process_Alley()
```

```python
def process_generic_category(name):
    global combined
    dummydata = pd.get_dummies(combined[name],prefix=name)
    combined = pd.concat([combined,dummydata],axis=1)
    combined.drop(name,axis=1,inplace=True)

def process_generic_numeric(name):
    global combined
    combined[name] = (combined[name] - combined[name].min())/(combined[name].max()-combined[name].min())


def process_generic_category_replace_nan_mode(name):
    global combined
    combined[name].fillna(combined[name].mode(),inplace=True)
    dummydata = pd.get_dummies(combined[name],prefix=name)
    combined = pd.concat([combined,dummydata],axis=1)
    combined.drop(name,axis=1,inplace=True)

def process_generic_numeric_replace_nan_mode(name):
    global combined
    combined[name].fillna(combined[name].mode(),inplace=True)    
    combined[name] = (combined[name] - combined[name].min())/(combined[name].max()-combined[name].min())


def process_generic_category_replace_nan_mean(name):
    global combined
    combined[name].fillna(combined[name].mean(),inplace=True)
    dummydata = pd.get_dummies(combined[name],prefix=name)
    combined = pd.concat([combined,dummydata],axis=1)
    combined.drop(name,axis=1,inplace=True)

def process_generic_numeric_replace_nan_mean(name):
    global combined
    combined[name].fillna(combined[name].mean(),inplace=True)    
    combined[name] = (combined[name] - combined[name].min())/(combined[name].max()-combined[name].min())

def process_generic_category_replace_nan_median(name):
    global combined
    combined[name].fillna(combined[name].median(),inplace=True)
    dummydata = pd.get_dummies(combined[name],prefix=name)
    combined = pd.concat([combined,dummydata],axis=1)
    combined.drop(name,axis=1,inplace=True)

def process_generic_numeric_replace_nan_median(name):
    global combined
    combined[name].fillna(combined[name].mean(),inplace=True)    
    combined[name] = (combined[name] - combined[name].min())/(combined[name].max()-combined[name].min())
    
    
def process_generic_category_replace_nan_unknown(name):
    global combined
    combined[name].fillna('unknown',inplace=True)
    dummydata = pd.get_dummies(combined[name],prefix=name)
    combined = pd.concat([combined,dummydata],axis=1)
    combined.drop(name,axis=1,inplace=True)

def process_generic_numeric_replace_nan_unknown(name):
    global combined
    combined[name].fillna('unknown',inplace=True)    
    combined[name] = (combined[name] - combined[name].min())/(combined[name].max()-combined[name].min())
```

# Modelling

```python
X_train = combined.head(1460).as_matrix()
X_test = combined.iloc[1460:].as_matrix()
y_train = train_output
```

```python
#Creating a keras model
def createmodel():
    model = Sequential()
    model.add(Dense(100,input_dim=299,kernel_initializer='normal',activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(40,kernel_initializer='normal',activation='relu'))
    model.add(Dense(40,kernel_initializer='normal',activation='relu'))
    model.add(Dense(1,kernel_initializer='normal',))

    #compiling the model
    model.compile(loss='mean_squared_error',optimizer='adam')

    #Fitting the model
    model.fit(X_train,y_train,validation_split=0.050,epochs=200,batch_size=64,verbose=0)
    return model

model1 = createmodel()
model2 = createmodel()
model3 = createmodel()
model4 = createmodel()
model5 = createmodel()
```

```python
#Predicting the values
prediction = model1.predict(X_test)
prediction = [int(round(single)) for single in prediction]
#Convertin to int values
prediction = [int(round(single)) for single in prediction]

#Predicting the values
prediction2 = model2.predict(X_test)
prediction2 = [int(round(single)) for single in prediction2]
#Convertin to int values
prediction2 = [int(round(single)) for single in prediction2]

#Predicting the values
prediction3 = model3.predict(X_test)
prediction3 = [int(round(single)) for single in prediction3]
#Convertin to int values
prediction3 = [int(round(single)) for single in prediction3]

#Predicting the values
prediction4 = model4.predict(X_test)
prediction4 = [int(round(single)) for single in prediction4]
#Convertin to int values
prediction4 = [int(round(single)) for single in prediction4]

#Predicting the values
prediction5 = model5.predict(X_test)
prediction5 = [int(round(single)) for single in prediction5]
#Convertin to int values
prediction5 = [int(round(single)) for single in prediction5]


data = np.vstack((prediction,prediction2,prediction3,prediction4,prediction5))
data = np.transpose(data)

dataframe = pd.DataFrame(data)
modes = dataframe.median(axis=1).as_matrix()
modes = np.transpose(modes)

data = np.vstack((HouseId,modes))
data = np.transpose(data)
```

```python
#Converting to dataframe
dataframe = pd.DataFrame(data)
dataframe.columns = ['Id','SalePrice']
dataframe.Id = dataframe.Id.astype(int)
print dataframe[:10]
```
